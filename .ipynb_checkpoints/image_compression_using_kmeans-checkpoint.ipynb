```python
# Importing allowed packages for the assignment. 
import numpy as np
import pandas as pd
from PIL import Image
from matplotlib.pyplot import imshow
%matplotlib inline
import time
```

```python
#Convert image to 3D array
def img_to_array(path):
    #Reads image and returns an numpy array
    arr_img = np.array(Image.open(path), dtype='int32')
    return arr_img
```

```python
#Reshape 3D to 2D array
def reshape_img(arr_img):
    r, c, l = arr_img.shape
    reshape_img = np.reshape(arr_img, (r*c, l), order="C")
    return reshape_img
```

```python
#Kmeans algorithm implementation
def kmeans(cls_array,k,maxlimit):
    centers = np.random.uniform(low=0, high=255, size=(k,3))
    df = pd.DataFrame({'r':cls_array[:,0], 'b': cls_array[:,1],'g': cls_array[:,2]})
    cls = np.zeros((len(df),len(centers)))
    centers_old = 0
    count = 0
    while (~np.equal(centers_old,centers).all()) and (count < maxlimit):
        centers_old = centers.copy()
        for j in range(k):
            cls[:,j] = np.sqrt((df['r']-centers[j][0])**2 + (df['b']-centers[j][1])**2  + (df['g']-centers[j][2])**2)
        df['cluster']  = np.argmin(cls, axis=1)    
        for m in range(k):
            centers[m] = df[['r','b','g']][df['cluster']==m].mean().values
        centers[np.isnan(centers)] = 0.0
        count += 1
        print('iteration : ',count, ' in progress')
    return df,centers
```

```python
#Constructing the compressed image
def img_from_kmeans(df, centers, k):
    df_img = df.copy()
    for i in range(k):
        df_img.loc[df_img['cluster'] == i,['r','b','g']] = centers[i]
    img_clustered = df_img[['r','b','g']].to_numpy()
    r, c, l = arr_img.shape
    reshape_img_clustered = np.reshape(img_clustered, (r,c,l), order="C")
    return reshape_img_clustered
```

```python
#Visualize the image
def array_to_img(arr):
    #Reads array and returns image
    arr = arr.astype(dtype='uint8')
    img = Image.fromarray(arr, 'RGB')
    return img
```

### 1. Clustering the beach.bmp image

```python
#Input-1 --> Read the beach.bmp image as an array and display it
arr_img = img_to_array("./data/beach.bmp")
print("Original image of beach.bmp :")
img = array_to_img(arr_img)
imshow(np.asarray(img))
```

```python
#Input-2 --> Define cluster size
k = 2

#Reshape the image from 3D to 2D 
res_img = reshape_img(arr_img)

#Run the kmeans algorithm and store the resultant dataframe and converging centroids
start = time.time()
df,centers = kmeans(res_img,k,300)
end = time.time()
print("Execution time: {} seconds ".format(end - start))

#Print the print(df.head()) which shows the r,g,b and its cluster assignments
print('\n')
print(df.head())
print('\n')

#Store the dataframe as a numpy vector
class_arr = df.cluster.to_numpy()

#Output#1 --> Convert the row vector to a column vector
col_class_arr = np.reshape(class_arr,(len(res_img), 1))
print("Cluster groupings as a column vector of rows {} and column {}".format(len(res_img),1))
print(col_class_arr)
print('\n')

#Output@2 --> Converging centroids
print("Centroids : ") 
print(centers)
print('\n')

#Reconstructing the clustered image 
reshape_img_clstr = img_from_kmeans(df, centers, k)
print("Clustered image with cluster of {} :".format(k))
img = array_to_img(reshape_img_clstr)

#Save image
img.save("./data/beach-k2.bmp")

#display image
imshow(np.asarray(img))
```
